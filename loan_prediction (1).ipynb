```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#creating a list consisting of only categorical features
# Initialize an empty list to store categorical features
categorical_features = []

# Iterate through each column and check if its datatype is 'object'
for feature in df.columns:
    if df[feature].dtypes == 'object':
        categorical_features.append(feature)

# Display the list of categorical features
print("Categorical Features:", categorical_features)
```

```python
numerical_features = []
for feature in df.columns:
    if df[feature].dtypes!='object':
        numerical_features.append(feature)
print("Numerical Features:",numerical_features)
```

```python
Continous_features=[]
Discrete_features=[]
for feature in numerical_features:
    print(f'Number of unique values in column {feature} is {df[feature].nunique()}')
    if df[feature].nunique()>15:
        Continous_features.append(feature)
    else:
        Discrete_features.append(feature)
```

```python
print("Continous_features:",Continous_features)
print("Discrete_features:",Discrete_features)
```

```python
for feature in categorical_features+Discrete_features:
    df[feature].fillna(df[feature].mode()[0],inplace=True)
```

```python
for feature in Continous_features:
    df[feature].fillna(df[feature].median(),inplace=True)
```

```python
df.head()
```

```python
for feature in df.columns:
    if(df[feature].nunique()>3):
        pass
    else:
        for i in range(df[feature].nunique()-1):
            df[feature]=np.where(str(df[feature].unique()[i]) == df[feature],1,0)
```

```python
df.head()
df = df.drop(['Loan_ID','Dependents'],axis=1)
```

```python
x = df.iloc[:,:10]
y=df['Loan_Status']
x
```

```python
scaled_data=x.copy()
```

```python
#splitting the dataset into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
```

```python
scaled_df=pd.DataFrame(min_max_scaler.fit_transform(scaled_data))
```

```python
scaled_df.head()
```

```python
x['ApplicantIncome']=scaled_df[4]
x['CoapplicantIncome']=scaled_df[5]
x['LoanAmount']=scaled_df[6]
x['Loan_Amount_Term']=scaled_df[7]
```

```python
x
```

```python
from sklearn.linear_model import LogisticRegression
#using logistic regression model
loreg = LogisticRegression()

loreg.fit(x_train, y_train)
y_pred=loreg.predict(x_test)
```

```python
print("Train Accuracy: ", loreg.score(x_test, y_test) * 100,' %')
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(loreg, x, y, cv=5)
print("Cross validation is",np.mean(score)*100)
```
